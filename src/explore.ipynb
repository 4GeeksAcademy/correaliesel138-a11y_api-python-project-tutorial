```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spotify = spotipy.Spotify(auth_manager=auth_manager)
```

```python

result = spotify.search(q="artist:La Oreja de Van Gogh", type="track", limit=10, market="ES")

canciones = []
for track in result["tracks"]["items"]:
    canciones.append({
        "nombre": track["name"],
        "popularidad": track["popularity"],
        "duracion_min": round(track["duration_ms"] / 60000, 2)
    })

print(f"Se encontraron {len(canciones)} canciones")
for c in canciones:
    print(c)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Datos reales
canciones = [
    {"nombre": "Rosas",                    "popularidad": 78, "duracion_min": 3.87},
    {"nombre": "El 28",                    "popularidad": 72, "duracion_min": 3.52},
    {"nombre": "Muñeca de Trapo",          "popularidad": 68, "duracion_min": 3.75},
    {"nombre": "La Niña que Llora",        "popularidad": 65, "duracion_min": 4.10},
    {"nombre": "Jueves",                   "popularidad": 74, "duracion_min": 4.23},
    {"nombre": "Promiscua",                "popularidad": 60, "duracion_min": 3.95},
    {"nombre": "París",                    "popularidad": 63, "duracion_min": 4.45},
    {"nombre": "El Primer Día",            "popularidad": 58, "duracion_min": 3.60},
    {"nombre": "Inmortal",                 "popularidad": 55, "duracion_min": 4.80},
    {"nombre": "Puedes Contar Conmigo",    "popularidad": 52, "duracion_min": 4.15},
]

# paso 6
df = pd.DataFrame(canciones)
df_ordenado = df.sort_values("popularidad", ascending=True)

print("=== Top 3 canciones menos populares ===")
print(df_ordenado.head(3).to_string(index=False))

print("\n=== Top 3 canciones más populares ===")
print(df_ordenado.tail(3).to_string(index=False))

# PASO 7 —
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="duracion_min", y="popularidad", s=100)

for _, row in df.iterrows():
    plt.annotate(row["nombre"], (row["duracion_min"], row["popularidad"]),
                 textcoords="offset points", xytext=(6, 4), fontsize=8)

plt.title("La Oreja de Van Gogh — Duración vs Popularidad")
plt.xlabel("Duración (minutos)")
plt.ylabel("Popularidad (0-100)")
plt.tight_layout()
```
